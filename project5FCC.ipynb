Use Pandas to import the data from epa-sea-level.csv.


```python
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\sunny\Downloads\flat-ui__data-Sat Dec 21 2024 (1).csv")
df
```

Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

```python

# Create the scatter plot
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], color='blue')

# Set the title and labels
ax.set_title('Rise in Sea Level')
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')

# Save the plot
fig.savefig('scatter_plot.png')

# Show the plot
plt.show()
```

Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

```python
from scipy.stats import linregress

def draw_plot():
    

    # Calculate the line of best fit
    slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

    # Create the line of best fit
    years_extended = pd.Series(range(1880, 2051))
    sea_levels_predicted = intercept + slope * years_extended
    ax.plot(years_extended, sea_levels_predicted, color='red', label='Best Fit Line')

    # Set the title and labels
    ax.set_title('Rise in Sea Level')
    ax.set_xlabel('Year')
    ax.set_ylabel('Sea Level (inches)')
    ax.legend()

    # Save the plot
    fig.savefig('scatter_plot_with_best_fit.png')

    # Show the plot
    plt.show()

    # Return the figure
    return fig

# Call the function to test it
draw_plot()

```

Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000

```python
def draw_plot():


    # Calculate the first line of best fit (using all data)
    slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
    years_extended = pd.Series(range(1880, 2051))
    sea_levels_predicted = intercept + slope * years_extended
    ax.plot(years_extended, sea_levels_predicted, color='red', label='Best Fit Line (All Data)')

    # Calculate the second line of best fit (using data from 2000 onwards)
    df_recent = df[df['Year'] >= 2000]
    slope_recent, intercept_recent, r_value_recent, p_value_recent, std_err_recent = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])
    sea_levels_predicted_recent = intercept_recent + slope_recent * years_extended
    ax.plot(years_extended, sea_levels_predicted_recent, color='green', label='Best Fit Line (2000 Onwards)')

    # Set the title and labels
    ax.set_title('Rise in Sea Level')
    ax.set_xlabel('Year')
    ax.set_ylabel('Sea Level (inches)')
    ax.legend()

    # Save the plot
    fig.savefig('scatter_plot_with_best_fit.png')

    # Show the plot
    plt.show()

    # Return the figure
    return fig

# Call the function to test it
draw_plot()
```
